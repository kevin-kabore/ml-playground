## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter (inline)
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # figsize=(width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

# 5. Save & show (you save the whole figure)
fig.savefig('images/sample-plot.png')
```

## Making figures with Numpy arrays

Baseline plots/figures:
* Line Plot
* Scatter Plot
* Bar plot
* Histogram
* Subplots

```python
# 1. Prepare data
# 2. Setup plot
# 3. Plot data
# 4. Customize plot
# 5. Save & show (you save the whole figure)
```

```python
# 1. Prepare data
x = np.linspace(0, 10, 100)
x[:10]
```

### Line plot

```python
# 2. Setup figure & 3. plot data
fig, ax = plt.subplots()
ax.plot(x, x**2); # line plot
```

### Scatter plot

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x)); # scatter plot
```

```python
# Anoter scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

### Bar plot from dict

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
x = nut_butter_prices.keys()
ax.bar(x, height=nut_butter_prices.values())

# 4. Customize plot
ax.set(title='The Nut Butter Store',
      xlabel='Type',
      ylabel='Price ($)');
```

```python
# 5. Save & show (you save the whole figure)
fig.savefig('images/nut-butter-bar.png')
```

```python
# horizontal bar graph .barh
# Must be turned into a list
fig, ax = plt.subplots()
y = nut_butter_prices.keys()
w = nut_butter_prices.values()
ax.barh(list(y), width=list(w));
```

### Histogram

```python
# Make some data for histograms and plot it
x = np.random.randn(1000) # random, normal distribution
x[:10]
```

```python
fig, ax = plt.subplots();
ax.hist(x);
```

### Subplots: two options

#### Subplot option 1

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, 
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axis
ax0.plot(x, x/2)
ax1.scatter(np.random.random(10), np.random.random(10))
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.hist(np.random.randn(1000));
```

#### Subplot option 2

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Plotting from pandas DataFrames

```python
import pandas as pd
```

```python
# Make a dataFrame
car_sales = pd.read_csv('car-sales.csv')
car_sales
```

```python
# Plotting from pandas dataFrame
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2020', periods=1000))
ts = ts.cumsum()
ts.plot()
```

### Plotting car_sales dataFrame

```python
## Clean up price
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '') # replace $ , . with empty string
car_sales
```

```python
type(car_sales['Price'][0])
```

```python
# Remove last two  zeroes
car_sales['Price'] = car_sales['Price'].str[:-2] # remove last two 
car_sales
```

```python
# Create a Sale Date column using pandas data range
car_sales['Sale Date'] = pd.date_range('1/1/2020', periods=len(car_sales))
car_sales
```

```python
# Make a total column using pd.Series.cumsum
# DOES NOT WORK: car_sales['Total Sales'] = car_sales['Price'].cumsum => type of Price is string
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

### Plot Total Sales x Sale Date

```python
car_sales.plot(x='Sale Date', y='Total Sales');
```
