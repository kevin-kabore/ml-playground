## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter (inline)
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # figsize=(width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

# 5. Save & show (you save the whole figure)
fig.savefig('images/sample-plot.png')
```

### Bar plot from dict

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
x = nut_butter_prices.keys()
ax.bar(x, height=nut_butter_prices.values())

# 4. Customize plot
ax.set(title='The Nut Butter Store',
      xlabel='Type',
      ylabel='Price ($)');
```

```python
# 5. Save & show (you save the whole figure)
fig.savefig('images/nut-butter-bar.png')
```

```python
# horizontal bar graph .barh
# Must be turned into a list
fig, ax = plt.subplots()
y = nut_butter_prices.keys()
w = nut_butter_prices.values()
ax.barh(list(y), width=list(w));
```

### Subplots: two options

#### Subplot option 1

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, 
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axis
ax0.plot(x, x/2)
ax1.scatter(np.random.random(10), np.random.random(10))
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.hist(np.random.randn(1000));
```

#### Subplot option 2

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Pyplot vs Matplotlin OO method
- When plotting something quickly, okay to use pyplot method (.plot)
- When plotting something more advanced, use OO method (Obeject Oriented)

#### Create a subset of the df for patients over 50 

```python
over_50 = heart_disease[heart_disease['age'] > 50];
over_50.head()
```

#### Create a scatter plot of the age and cholestorol columns, colored with target col

```python
# pyplot method - Eh example
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target', # c = color
             figsize=(10,6)); 
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

# set x axis to be greater
#ax.set_xlim([45, 100]);
```

```python
# Object Oriented method from scratch
fig, ax = plt.subplots(figsize=(10,6)) # 1. create figure
# 2. plot the data
scatter = ax.scatter(x=over_50['age'],
           y=over_50['chol'],
           c=over_50['target'])
# Customize the plot
ax.set(title='Heart Disease and Cholestorol levels',
       xlabel='Age',
       ylabel='Cholesterol')

# Add a legend
# grabs all the values and creates a legend from the c param in scatter
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line to show the mean
ax.axhline(over_50['chol'].mean(),
           )
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               sharex=True,
                               figsize=(10,10))
# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
           y=over_50['chol'],
           c=over_50['target'])

# Customize the plot
ax0.set(title='Heart Disease and Cholestorol Levels',
        xlabel='Age',
        ylabel='Cholesterol')

# Add a legend
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a mean line
ax0.axhline(y=over_50['chol'].mean(),
            color='r',
            label='Average',
            linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
            y=over_50['thalach'],
            c=over_50['target']);

# Customize the plot
ax1.set(title='Heart Disease and Max Heart Rate Levels',
        xlabel='Age',
        ylabel='Max Heart Rate')

# Add a legend
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a mean line
ax1.axhline(y=over_50['thalach'].mean(),
            color='r',
            label='Average',
            linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
# Create a pyplot from the pandas df
ax = df.plot(kind='bar')
# Customize the plot with the set() method
ax.set(title='Random Number Bar Graph from DataFrame',
       xlabel='Row Number',
       ylabel='Random Number')
# Make the legend visible
ax.legend().set_visible(True)
```

```python
plt.style.use('seaborn-whitegrid')
# Object Oriented method from scratch
fig, ax = plt.subplots(figsize=(10,6)) # 1. create figure

# 2. plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter') # this changes the color scheme

# Customize the plot
ax.set(title='Heart Disease and Cholestorol levels',
       xlabel='Age',
       ylabel='Cholesterol')

# Add a legend
# grabs all the values and creates a legend from the c param in scatter
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line to show the mean
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
# Customize the x and y axis limitations
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               sharex=True,
                               figsize=(10,10))
# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter')

# Customize the plot
ax0.set(title='Heart Disease and Cholestorol Levels',
        xlabel='Age',
        ylabel='Cholesterol')
# change the x axis limits
ax0.set_xlim([50, 80])

# Add a legend
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a mean line
ax0.axhline(y=over_50['chol'].mean(),
            color='r',
            label='Average',
            linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='winter');

# Customize the plot
ax1.set(title='Heart Disease and Max Heart Rate Levels',
        xlabel='Age',
        ylabel='Max Heart Rate')
# change the x axis limits on ax1
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a mean line
ax1.axhline(y=over_50['thalach'].mean(),
            color='r',
            label='Average',
            linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
